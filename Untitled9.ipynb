```python
import numpy as np
import matplotlib.pyplot as plt

# Diode parameters
I_s = 1e-12  # Reverse saturation current in A
V_t = 0.025  # Thermal voltage in V (approx. 25mV at room temperature)

def diode_current(V):
    """Calculate the current through the diode for a given voltage V."""
    return I_s * (np.exp(V / V_t) - 1)

# Voltage range for the I-V characteristic
V = np.linspace(-0.7, 0.7, 400)
I = diode_current(V)

# Plotting the I-V characteristics
plt.figure(figsize=(10, 5))
plt.plot(V, I, label='Diode I-V Characteristic')
plt.title('Diode I-V Characteristics')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.ylim(-1e-3, 1e-3)
plt.xlim(-0.7, 0.7)
plt.grid()
plt.legend()
plt.show()
```
